```python
def add(a, b):
    return a - b

print(add(5, 3))
```

```python
numbers = [5, 8, 2, 10, 3]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print(largest)
```

```python
def calculate_average(numbers):
    if len(numbers) == 0:
        return 0

    total = 0
    for num in numbers:
        total += num

    return total / len(numbers)

print(calculate_average([10, 20, 30]))
```
