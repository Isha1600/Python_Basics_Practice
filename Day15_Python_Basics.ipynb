```python
students = [
    {"name": "Isha", "marks": 90},
    {"name": "Tina", "marks": 85}
]

# Function to save the entire list of students by overwriting the file
def overwrite_students_file(student_list):
    with open("students.txt", "w") as file:
        for student in student_list:
            line = student["name"] + "," + str(student["marks"]) + "\n"
            file.write(line)

# Function to add a single student by appending to the file
def add_student_to_file(name, marks):
    with open("students.txt", "a") as file:
        line = name + "," + str(marks) + "\n"
        file.write(line)

# Initial save of students using the overwrite function
overwrite_students_file(students)
```

```python
def load_students():
    student_list = []
    with open("students.txt", "r") as file:
        for line in file:
            name, marks = line.strip().split(",")
            student = {"name": name, "marks": int(marks)}
            student_list.append(student)
    return student_list

students = load_students()
print(students)
```

```python
products = [
    {"name": "Laptop", "price": 1200},
    {"name": "Mouse", "price": 25}
]

def overwrite_products_file(product_list):
    with open("products.txt", "w") as file:
        for product in product_list:
            line = product["name"] + "," + str(product["price"]) + "\n"
            file.write(line)

def add_product_to_file(name, price):
    with open("products.txt", "a") as file:
        line = name + "," + str(price) + "\n"
        file.write(line)

overwrite_products_file(products)
```

```python
def load_products():
    product_list = []
    with open("products.txt", "r") as file:
        for line in file:
            name, price = line.strip().split(",")
            product = {"name": name, "price": float(price)}
            product_list.append(product)
    return product_list

products = load_products()
print(products)
```
