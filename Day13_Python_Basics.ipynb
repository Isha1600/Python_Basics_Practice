```python
students = {
    "Isha": {
        "age": 21,
        "marks": 90
    },
    "Tina": {
        "age": 22,
        "marks": 85
    },
    "Riya": {
        "age": 20,
        "marks": 88
    }
}
print(students["Tina"]["marks"])
```

```python
students = [
    {"name": "Isha", "age": 21, "marks": 90},
    {"name": "Tina", "age": 22, "marks": 85},
    {"name": "Riya", "age": 20, "marks": 88}
]
for student in students:
    print(student["name"], student["marks"])
```

```python
total = 0

for student in students:
    total += student["marks"]

average = total / len(students)
print("Average:", average)
```

```python
highest = 0
top_student = ""

for student in students:
    if student["marks"] > highest:
        highest = student["marks"]
        top_student = student["name"]

print("Top student:", top_student)
```

```python
products = [
    {"name": "Laptop", "price": 1200, "quantity": 5},
    {"name": "Mouse", "price": 25, "quantity": 20},
    {"name": "Keyboard", "price": 75, "quantity": 10},
    {"name": "Monitor", "price": 300, "quantity": 3}
]
```

```python
total_inventory_value = 0
for product in products:
    total_inventory_value += (product["price"] * product["quantity"])

print(f"Total Inventory Value: ${total_inventory_value}")
```

```python
most_expensive_product = None
highest_price = 0

for product in products:
    if product["price"] > highest_price:
        highest_price = product["price"]
        most_expensive_product = product

print(f"The most expensive product is: {most_expensive_product['name']} with a price of ${most_expensive_product['price']}")
```

```python
print("Students with marks greater than 85:")
for student_name, student_data in students.items():
    if student_data["marks"] > 85:
        print(student_name)
```
