```python
def print_numbers():
    for i in range(1, 6):
        print(i)

print_numbers()
```

```python
def print_my_name_five_times(name):
    for _ in range(5):
        print(name)

print_my_name_five_times("Colab User")
```

```python
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Test the function
print(f"5 is {check_even_odd(5)}")
print(f"10 is {check_even_odd(10)}")
```

```python
def multiply(a, b):
    return a * b

result_multiplication = multiply(7, 6)
print(result_multiplication)
```

```python
def square(number):
    return number * number

# Test the function
num = 7
squared_num = square(num)
print(f"The square of {num} is {squared_num}")
```

```python
def sum_list_numbers(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

# Test the function
my_list = [1, 2, 3, 4, 5]
list_sum = sum_list_numbers(my_list)
print(f"The sum of the list {my_list} is {list_sum}")
```
