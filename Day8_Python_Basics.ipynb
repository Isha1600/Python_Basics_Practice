```python
def count_vowels(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
    return count

print(count_vowels("Isha"))
```

```python
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

```python
def is_palindrome(text):
    if text == text[::-1]:
        return True
    else:
        return False

print(is_palindrome("madam"))
```

```python
def count_greater_than_10(numbers):
    count = 0
    for number in numbers:
        if number > 10:
            count += 1
    return count

my_list = [10, 5, 20, 15, 30, 8, 25]
print(f"The number of elements greater than 10 in the list {my_list} is: {count_greater_than_10(my_list)}")
```

```python
def find_largest_number(numbers):
    if not numbers:
        return None  # Handle empty list case
    largest = numbers[0]
    for number in numbers:
        if number > largest:
            largest = number
    return largest

my_list = [10, 5, 20, 15, 30, 25]
print(f"The largest number in the list {my_list} is: {find_largest_number(my_list)}")
```

```python
def count_consonants(text):
    count = 0
    vowels = "aeiou"
    for char in text.lower():
        if char.isalpha() and char not in vowels:
            count += 1
    return count

print(f"Number of consonants in 'Python': {count_consonants('Python')}")
print(f"Number of consonants in 'Hello World': {count_consonants('Hello World')}")
print(f"Number of consonants in 'AEIOU': {count_consonants('AEIOU')}")
```

```python
def remove_spaces(text):
    return text.replace(" ", "")

input_string = "Data Science"
result_string = remove_spaces(input_string)
print(f"Original string: '{input_string}'")
print(f"String with spaces removed: '{result_string}'")
```

```python
def count_letter_a(text):
    count = 0
    for char in text.lower():
        if char == 'a':
            count += 1
    return count

print(f"Number of 'a's in 'banana': {count_letter_a('banana')}")
print(f"Number of 'a's in 'Apple and Orange': {count_letter_a('Apple and Orange')}")
print(f"Number of 'a's in 'Python': {count_letter_a('Python')}")
```

```python
def contains_only_digits(text):
    return text.isdigit()

print(f"'12345' -> {contains_only_digits('12345')}")
print(f"'12a45' -> {contains_only_digits('12a45')}")
print(f"'' -> {contains_only_digits('')}") # Test with an empty string
```

```python
def find_longest_word(sentence):
    words = sentence.split()
    if not words:
        return "" # Return empty string for an empty sentence

    longest_word = ""
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word
    return longest_word

input_sentence = "I love learning Python programming"
result_word = find_longest_word(input_sentence)
print(f"The longest word in '{input_sentence}' is: '{result_word}'")
```

```python
def reverse_string_loop(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    return reversed_text

print(f"'Python' reversed is: '{reverse_string_loop('Python')}'")
print(f"'Hello' reversed is: '{reverse_string_loop('Hello')}'")
print(f"'a' reversed is: '{reverse_string_loop('a')}'")
print(f"'' reversed is: '{reverse_string_loop('')}'")
```
