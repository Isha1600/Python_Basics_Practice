```python
student = {
    "Isha": 90,
    "Tina": 85,
    "Riya": 88
}

highest = 0
top_student = ""

for name, marks in student.items():
    if marks > highest:
        highest = marks
        top_student = name

print("Top student:", top_student)
```

```python
students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}
print("Student dictionary created.")
```

```python
print("Student Names:")
for name in students:
    print(name)
```

```python
print("Student Marks:")
for marks in students.values():
    print(marks)
```

```python
total_marks = sum(students.values())
num_students = len(students)

if num_students > 0:
    average_marks = total_marks / num_students
    print(f"Average Marks: {average_marks:.2f}")
else:
    print("No students in the dictionary to calculate average marks.")
```

```python
lowest_mark = float('inf') # Initialize with a very high value
lowest_student = ""

for name, marks in students.items():
    if marks < lowest_mark:
        lowest_mark = marks
        lowest_student = name

print(f"Student with the lowest marks: {lowest_student} ({lowest_mark})")
```

```python
students_above_80 = 0

for marks in students.values():
    if marks > 80:
        students_above_80 += 1

print(f"Number of students who scored above 80: {students_above_80}")
```
