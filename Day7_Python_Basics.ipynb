```python
numbers = [10, 20, 30]

print(numbers[0])  # 10
print(numbers[1])  # 20
print(numbers[2])  # 30
```

```python
numbers = [2, 4, 6, 8]

for n in numbers:
    print(n)
```

```python
def print_even(numbers):
    for n in numbers:
        if n % 2 == 0:
            print(n)

nums = [1,2,3,4,5,6,7,8]
print_even(nums)
```

```python
nums = [1,2,3]
nums.append(4)
print(nums)
nums.remove(2)
print(nums)
```

```python
def analyze_marks(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)
    print("Total:", total)
    print("Average:", average)

student_marks = [80, 75, 90, 85]
analyze_marks(student_marks)
```
