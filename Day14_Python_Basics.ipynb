```python
def calculate_average(student_list):
    total = 0
    for student in student_list:
        total += student["marks"]
    return total / len(student_list)


def find_top_student(student_list):
    highest = 0
    top_name = ""
    for student in student_list:
        if student["marks"] > highest:
            highest = student["marks"]
            top_name = student["name"]
    return top_name


students = [
    {"name": "Isha", "marks": 90},
    {"name": "Tina", "marks": 85},
    {"name": "Riya", "marks": 95}
]

print("Average:", calculate_average(students))
print("Top student:", find_top_student(students))
```

```python
students = []

def add_student(name, marks):
    student = {"name": name, "marks": marks}
    students.append(student)

def show_students():
    for student in students:
        print(student["name"], student["marks"])

def find_top_student():
    highest = 0
    top_name = ""
    for student in students:
        if student["marks"] > highest:
            highest = student["marks"]
            top_name = student["name"]
    return top_name

def remove_student(name):
    global students # Declare students as global to modify it
    initial_len = len(students)
    students = [student for student in students if student["name"] != name]
    if len(students) < initial_len:
        print(f"Student {name} removed successfully.")
    else:
        print(f"Student {name} not found.")


# Using the system
add_student("Isha", 90)
add_student("Tina", 85)
add_student("Riya", 95)

print("Students before removal:")
show_students()
print("Top student before removal:", find_top_student())

print("\n--- Removing Tina ---")
remove_student("Tina")

print("\nStudents after removal:")
show_students()
print("Top student after removal:", find_top_student())

print("\n--- Trying to remove a non-existent student ---")
remove_student("John")
```

```python
def students_above_80(student_list):
    above_80_students = []
    for student in student_list:
        if student["marks"] > 80:
            above_80_students.append(student["name"])
    return above_80_students

students = [
    {"name": "Isha", "marks": 90},
    {"name": "Tina", "marks": 85},
    {"name": "Riya", "marks": 78},
    {"name": "Sam", "marks": 92}
]

print("Students scoring above 80:", students_above_80(students))
```

```python
def lowest_marks(student_list):
    if not student_list:
        return None  # Handle empty list case
    lowest = student_list[0]["marks"]
    for student in student_list:
        if student["marks"] < lowest:
            lowest = student["marks"]
    return lowest

students = [
    {"name": "Isha", "marks": 90},
    {"name": "Tina", "marks": 85},
    {"name": "Riya", "marks": 78},
    {"name": "Sam", "marks": 92}
]

print("Lowest marks:", lowest_marks(students))
```
