```python
try:
    number = int(input("Enter number: "))
    print(10 / number)
except:
    print("Something went wrong!")
```

```python
try:
    number = int(input("Enter number: "))
    print(10 / number)
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("Please enter valid number!")
```

```python
try:
    number = int(input("Enter number: "))
    result = 10 / number
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Result is:", result)
```
