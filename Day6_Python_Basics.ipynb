```python
secret = 7
guess = 0

while guess != secret:
    guess = int(input("Guess a number between 1 and 10: "))
    if guess == secret:
        print("Correct! ðŸŽ‰")
    else:
        print("Try again!")
```

```python
num = 10
while num >= 1:
    print(num)
    num -= 1
```
